# NAIVE BAYES CLASSIFIER

**Reading the Dataset**

```python
import pandas as pd

df = pd.read_excel('DATASET.xlsx',index_col ='Example')
df
```

**Function for Generating the Probability Table**

```python
import numpy as np

def get_prob_table(df,label):
    prob_table = {}
    values,counts = np.unique(df[label],return_counts = True)
    idx = {values[i]:i for i in range(len(values))}
    for col in df.columns[:-1]:
        table = {}
        for x,y in zip(values,counts):
            dff = df.where(df[label] == x).dropna()
            val,cnt = np.unique(dff[col],return_counts = True)
            for a,b in zip(val,cnt):
                if a not in table:
                    table[a] = np.ones(len(values))*0.001 #Setting the minimum probability to 10e-3
                table[a][idx[x]] = b/y
        table[col] = values
        prob_table[col] = pd.DataFrame(table).set_index(col).T
    prob = {}
    for i in range(len(values)):
        prob[values[i]] = counts[i]/len(df)
    return prob_table,pd.DataFrame(prob,index = ['Probability'])
```

**Function to Classify a Data Point**

```python
def classify(prob_table,prob,X):
    probab = {}
    for x in prob:
        pr = prob[x]['Probability']
        for var in X:
            m = prob_table[var]
            pr = pr*m[x][X[var]]
        probab[x] = pr
    return max(prob,key = probab.get)
```

***PROBABILITY TABLE***

```python
label = 'Target Wait'
prob_table, prob = get_prob_table(df,label)
for x in prob_table:
    print('Probability Table of',x)
    print(prob_table[x],'\n')
print('Probability of values in the Target Column')
print(prob)
```

***CLASSIFICATION***

```python
X = {'ALT': 'T', 
     'Bar': 'T', 
     'Fri': 'F', 
     'Hun': 'T', 
     'Pat': 'Some', 
     'Price': '$$$', 
     'Rain': 'T', 
     'Res': 'T', 
     'Type': 'Italian', 
     'Est': '> 60'}
```

```python
a = classify(prob_table,prob,X)
print(f'The given Data Point is classified as {label} = {a}')
```

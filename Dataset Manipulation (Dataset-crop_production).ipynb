**READING THE DATASET**

```python
import pandas as pd
df = pd.read_csv('crop_production.csv')
df.head()
```

**TRANSFORMING THE GIVEN DATASET**

```python
dff = pd.pivot_table(df, index=['State_Name', 'District_Name','Season','Crop'],columns = ['Crop_Year'],values = ['Area','Production'])
dff = dff.swaplevel(0,1, axis = 1)
dff = dff.sort_index(axis = 1)
dff
```

**SLICING THE TRANSFORMED DATASET**

```python
df1 = dff[:len(dff)//2]
df2 = dff[len(dff)//2:]
```

```python
df1
```

```python
df2
```

**COUNTING NaN VALUES FEATURE WISE**

```python
print('\tNaN in DF1\n')
print('Feature:\t\tCount')
i = 0
for col in df1.iteritems():
    if i == 0:
        print(f'{col[0]}:\t\t{df1[col[0]].isna().sum()}')
        i = 1
    else:
        print(f'{col[0]}:\t{df1[col[0]].isna().sum()}')
        i = 0
```

```python
print('\tNaN in DF2\n')
print('Feature:\t\tCount')
i = 0
for col in df2.iteritems():
    if i == 0:
        print(f'{col[0]}:\t\t{df2[col[0]].isna().sum()}')
        i = 1
    else:
        print(f'{col[0]}:\t{df2[col[0]].isna().sum()}')
        i = 0
```

**USING ffill() TO REPLACE NaN VALUES IN DF1**

```python
dff1 = df1.fillna(method = 'ffill')
dff1
```

**USING bfill() TO REPLACE NaN VALUES IN DF2**

```python
dfb2 = df2.fillna(method = 'bfill')
dfb2
```

**REPLACING NaN WITH MEAN IN DF1**

```python
dfm1 = dff1.fillna(value = dff1.mean())
dfm1
```

**REPLACING NaN WITH MEDIAN IN DF2**

```python
dfM2 = dfb2.fillna(value = dfb2.median())
dfM2
```

```python
DD1 = dfm1.fillna(0)
DD2 = dfM2
```

**FINDING CORRELATION AND VISUALISATION USING HEATMAP**

```python
#Correlation of df1
CD1 = DD1.corr()
#Correlation of df2
CD2 = DD2.corr()
```

**CONVERT df1 INTO MIN-MAX NORMALISED FORMAT**

```python
from sklearn.preprocessing import MinMaxScaler
trans_df1 = MinMaxScaler().fit_transform(DD1)
pd.DataFrame(trans_df1)
```

```python
from sklearn.preprocessing import StandardScaler
trans_df2 = StandardScaler().fit_transform(DD2)
pd.DataFrame(trans_df2)
```
